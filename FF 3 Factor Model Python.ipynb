```python
class famafrench():
    def __init__(self, y, start = '1963-7-1', end = '1990-12-1'):
        self.start = start
        self.end = end
        self.y = y[start:end]
    
    def __str__(self):
        return 'from %s to %s' %(self.start, self.end)

    def variables(self):
        x = ff3[self.start:self.end]
        y = self.y[self.start:self.end]
        return x, y

    def regression(self):
        x = self.variables()[0]
        x = sm.add_constant(x)
        y = self.variables()[1]
        model = sm.OLS(y,x)
        result = model.fit()

        return result
        

```
