```python
import statsmodels.api as sm #for OLS regression
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt #for heatmap plotting
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns #for heatmap plotting
from statsmodels.iolib.summary2 import summary_col #to summarize multiple regressions in one table
from datetime import date
from dateutil.relativedelta import relativedelta
import math
from scipy import stats
```

```python
Stock_Quantity = pd.read_csv('C:/Users/yz2991/Desktop/Stock Quantity in Size Portfolio.csv')
```

```python
Stock_Quantity.rename(columns = {'Unnamed: 0': 'YYMM'}, inplace = True)
Stock_Quantity['YYMM'] = pd.to_datetime(Stock_Quantity['YYMM'],yearfirst=True, format='%Y%m')
```

```python
Stock_Quantity.index = Stock_Quantity['YYMM']
Stock_Quantity = Stock_Quantity.drop('YYMM', axis = 1)
```

```python
print(Stock_Quantity.head)
```

```python
fig=plt.figure(figsize=(18, 6), dpi= 80, facecolor='w', edgecolor='k')
pal = sns.color_palette("Set1")
plt.stackplot(Stock_Quantity.index,Stock_Quantity['Lo 30'], Stock_Quantity['Med 40'], Stock_Quantity['Hi 30'], labels=['Lo 30','Med 40','Hi 30'],colors=pal, alpha=0.5)

plt.xticks(rotation=90)
plt.xlabel('Time', size = 10)
plt.ylabel('Number of Firms in Portfolios', size = 10, rotation = 90)
plt.title('Number of Firms in Portfolios Sorted in Size', size = 13)
plt.xticks(size = 10)
plt.legend(loc='upper left')
plt.show()
```
